```python
import pandas as pd

# Loading the dataset to randomly sample the datsets and analyse the classification of Dietgroups!
df = pd.read_csv("Visualization_Dataset.csv")

# Renaming important columns 
df.rename(columns={
    'diet_group': 'Diet',
    'sex': 'Gender',
    'age_group': 'AgeGroup',
    'mean_ghgs': 'GHG_Total',
    'mean_ghgs_ch4': 'CH4',
    'mean_ghgs_n2o': 'N2O',
    'mean_land': 'LandUse',
    'mean_watuse': 'WaterUse',
    'mean_watscar': 'WaterScarcity',
    'mean_eut': 'Eutrophication',
    'mean_bio': 'Biodiversity',
    'mean_acid': 'Acidification'
}, inplace=True)

# Picking only the important broad columns for refering to the TreeMap!
df_sample = df[['Gender', 'AgeGroup', 'Diet', 'GHG_Total', 'LandUse', 'WaterUse']]

# Randomly sample 15 rows to visualize
sampled_df = df_sample.sample(n=15, random_state=42)


display(HTML("<h2 style='color: navy;'> Visualizing Sampled Dataset </h2>"))
# Visualize the sampled rows
sampled_df
```

```python
import plotly.express as px
import pandas as pd

def plot_treemap(feature_column, title_text):
    fig = px.treemap(
        df_modified,
        path=['Gender', 'Diet', 'AgeGroup'],
        values=feature_column,
        color=feature_column,
        color_continuous_scale='Blues',
        title=title_text,
        width=1000,
        height=700
    )
    fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to plot the boxplot for any chosen feature and visualize any outliers if present!!
def plot_boxplot(feature_column):
    plt.figure(figsize=(14, 6))
    sns.boxplot(data=df_modified, x='Diet',hue='Diet', y=feature_column, palette="Set2",legend=False)
    plt.title(f'{feature_column} by Diet Type', fontsize=16)
    plt.xlabel('Diet Type', fontsize=12)
    plt.ylabel(feature_column, fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()

# Interactive choice
# List of features a user can choose from!!
features = [
    'GHG_Total', 'LandUse', 'WaterScarcity', 'Eutrophication',
    'CH4', 'N2O', 'Biodiversity', 'WaterUse', 'Acidification'
]

print("Choose one feature to visualize from below list:\n")
for idx, feat in enumerate(features):
    print(f"{idx + 1}. {feat}")

# Take user input
choice = int(input("\nEnter the number corresponding to the feature: "))

# Condition for Validating choice and generating corresponding plot!!
if 1 <= choice <= len(features):
    selected_feature = features[choice - 1]
    plot_boxplot(selected_feature)
else:
    print("Invalid choice. Please select a valid number!")
```
